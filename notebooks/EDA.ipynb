```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Defining basic functions

def read_file(filename):
    df= pd.read_csv(filename)
    return df

def test_run(input_dataframe):
    df = input_dataframe
    print ("data frame head")
    print (df.head())
    print ("data frame tail")
    print (df.tail())
    print ("data frame volume")
    print (df["Volume"])
    
def test_plot(input_dataframe):
    df = input_dataframe
    print ("Univariate plot")
    df['Close'].plot()
    plt.show()

def test_multiplot(input_dataframe):
    df = input_dataframe
    print ("multivariate plot")
    df[['Open','Close']].plot()
    plt.show()
```

```python
if __name__=="__main__":
    file_path= "../data/raw/CAT.csv"
    data = read_file(file_path)
    test_multiplot(data)
    test_run(data)
    test_plot(data)
```
